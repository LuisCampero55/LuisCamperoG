```
import matplotlib.pyplot as plt
import math

def e1(y):
    return (.25 + (.25*y))
def e2(x, z):
  return ((.25) + (.25*x) + (.25*z))
def e3(y, w):
    return ((.25) + (.25*y) + (.25*w))
def e4(z):
    return (.25 + (.25*z))

#Nuevo método
def NuevoMetodo():
  print("Método nuevo")
  Mfg1()

def Mfg1():
  xDat=[]
  xDat.append(0)
  yDat=[]
  yDat.append(0)
  zDat=[]
  zDat.append(0)
  wDat=[]
  wDat.append(0)
  cont = 0
  x = 0
  y = 0
  z = 0
  w = 0
  #it = True
  while(cont < 11):
    xd = e1(y)
    xDat.append(xd)
    yd = e2(x , z)
    yDat.append(yd)
    zd = e3(y, w)
    zDat.append(zd)
    wd = e4(z)
    wDat.append(wd)
    x = xd
    y = yd
    z = zd
    w = wd
    cont += 1
  plt.plot(xDat, label = "x_Dato")
  plt.plot(yDat, label = "y_Dato")
  plt.plot(zDat, label = "z_Dato")
  plt.plot(wDat, label = "w_Dato")
  plt.legend()
  plt.show()
  print("Iteraciones: \n")
  print("x : " + str(xDat))
  print("y : " + str(yDat))
  print("z : " + str(zDat))
  print("w : " + str(wDat))

NuevoMetodo()
```
