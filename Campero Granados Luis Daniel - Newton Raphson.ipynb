```
#Método Newton Raphson 1

import matplotlib.pyplot as plt
import math

def NewtonRaphson():
  print("Método Newton Raphson")
  xi = int(input("Ingrese su valor inicial: "))
  cp = float(input("Ingrese su criterio de paro: "))
  fg(xi,cp)

def g1(x):
  return ((math.e**-x)-x)

def gg1(x):
  return ((-math.e**-x)-1)

def fg(xinicial, criterio=0.0001):
  x_vec=[]
  x_vec.append(xinicial)
  contador = 0
  error_vec=[]
  error_vec.append(1)
  criterio_paro = criterio
  it = True
  while(error_vec[contador] > criterio_paro):
    xx = g1(x_vec[contador])
    yy = gg1(x_vec[contador])
    div = (xx / yy) if yy != 0 else 0
    x_temp = (x_vec[contador]-div)
    x_vec.append(x_temp)
    e_temp = abs(x_vec[contador+1]-x_vec[contador])
    error_vec.append(e_temp)
    contador += 1
    if (x_vec[contador]>=2000):
      it = False
      break
  print("Raiz = "+str(x_vec[contador-1]) + "\n" + str(it))
  plt.plot(x_vec, label = "x_r")
  plt.plot(error_vec, label = "error")
  plt.legend()
  plt.show()  

NewtonRaphson()
```

```
#Método Newton Raphson 2

import matplotlib.pyplot as plt
import math

def NewtonRaphson2():
  print("Método Newton Raphson con la funcion: (2 * (x^2)) - (x - 5)")
  xi = int(input("Ingrese su valor inicial: "))
  cp = float(input("Ingrese su criterio de paro: "))
  fg(xi,cp)

def g1(x):
  return ((2*(x**2)-x-5))

def gg1(x):
  return ((4*x)-1)

def fg(xinicial, criterio=0.0001):
  x_vec=[]
  x_vec.append(xinicial)
  contador = 0
  error_vec=[]
  error_vec.append(1)
  criterio_paro = criterio
  it = True
  while(error_vec[contador] > criterio_paro):
    xx = g1(x_vec[contador])
    yy = gg1(x_vec[contador])
    div = (xx / yy) if yy != 0 else 0
    x_temp = (x_vec[contador]-div)
    x_vec.append(x_temp)
    e_temp = abs(x_vec[contador+1]-x_vec[contador])
    error_vec.append(e_temp)
    contador += 1
    if (x_vec[contador]>=2000):
      it = False
      break
  print("Raiz = "+str(x_vec[contador-1]) + "\n" + str(it))
  plt.plot(x_vec, label = "x_r")
  plt.plot(error_vec, label = "error")
  plt.legend()
  plt.show()  

NewtonRaphson2()
```
