```
import matplotlib.pyplot as plt
import math

#Ejercicio 1
def e11(y, z):
    return (-.222 - (.222*y) + (.111*z))
def e12(x, z):
    return ((.375) - (.875*x) - (.625*z))
def e13(x, y):
    return ((-.600) + (.300*x) + (.400*y))

#Ejercicio 2
def e21(y):
    return (2.4 - (.4*y))
def e22(x):
    return ((.8) + (.1*x))

#Ejercicio 3
def e31(y):
    return (.5 - (.125*y))
def e32(x):
    return ((.6) - (.4*x))
def e33(x):
    return ((.75) - (.25*x))

#Ejercicio 4
def e41(y, z):
    return (3.667 - (.333*y) - (.167*z))
def e42(x, z):
    return ((2.5) + (.125*x) - (.25*z))
def e43(x, y):
    return ((3.833) - (.167*x) + (.167*y))
```
